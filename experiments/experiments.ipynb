```python
import requests
from bs4 import BeautifulSoup

def fetch_boomlive_fact_check(query: str):
    search_url = f"https://www.boomlive.in/search?q={query}"
    response = requests.get(search_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the first fact-check article link
    article_link = soup.find('a', {'class': 'article-title'})
    if article_link:
        article_url = article_link['href']
        article_response = requests.get(article_url)
        article_soup = BeautifulSoup(article_response.text, 'html.parser')
        
        # Extract title and content
        title = article_soup.find('h1').get_text(strip=True)
        content = article_soup.find('div', {'class': 'article-content'}).get_text(strip=True)
        
        return {'title': title, 'content': content, 'url': article_url}
    else:
        return None

    

if __name__ == "__main__":
    query = "Modi"
    result = fetch_boomlive_fact_check(query)
    if result:
        print(f"Title: {result['title']}")
        print(f"URL: {result['url']}")
        print(f"Content Snippet: {result['content'][:300]}...")
    else:
        print("No fact-check article found.")

```

```python
from bs4 import BeautifulSoup
import requests

# URL for BOOMLive search results
url = "https://www.boomlive.in/search?q=Modi"

response = requests.get(url)

print(response.status_code)
print(response.headers['Content-Type'])

soup = BeautifulSoup(response.text, 'html.parser')

trending_tags_section = soup.find('div', {'class': 'trending-tags'}) 

if trending_tags_section:
    # Extract all the tags (they might be inside <a> or <span> tags, depending on the structure)
    tags = trending_tags_section.find_all(['a', 'span'])
    print("Trending Tags:")
    
    for tag in tags:
        print(tag.text.strip())
else:
    print("Trending Tags not found.")
```
